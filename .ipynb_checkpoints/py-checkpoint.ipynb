# Titanic

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split #Spliting data
from sklearn.linear_model import LogisticRegression #For Logistic Regression
from sklearn.metrics import classification_report,confusion_matrix #for Classification data
from sklearn.neighbors import KNeighborsClassifier # for KNN method
```

```python
train = pd.read_csv("titanic_train.csv")
```

```python
train.head()
```

```python
# heatmap of missing data
plt.figure(figsize=(18,6))
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap="viridis" )
```

# Data Cleaning

```python
plt.figure(figsize=(18,6))
sns.boxplot(x ="Pclass", y = "Age", data = train)
```

```python
# Defined the function for replacing the missing Age data by means based on Pclass
def impute_age(cols):
    
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        
        elif Pclass == 2:
            return 29
        
        else:
            return 24
    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
# checking the missing data
plt.figure(figsize=(18,6))
sns.heatmap(train.isnull(),yticklabels=False, cbar=False, cmap="viridis")
```

```python
# the missing data of cabin is too much it is better to drop this column
train.drop("Cabin", axis = 1, inplace = True)
```

```python
plt.figure(figsize=(18,6))
sns.heatmap(train.isnull(),yticklabels=False, cbar=False, cmap="viridis")
```

```python
train.dropna(inplace = True)
```

```python
plt.figure(figsize=(18,6))
sns.heatmap(train.isnull(),yticklabels=False, cbar=False, cmap="viridis")
```

# EDA

# Converting Categorical Features

```python
Sex = pd.get_dummies(train["Sex"], drop_first= True)
embark= pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train = pd.concat([train,Sex,embark], axis = 1)
```

```python
train.drop(["Name","Sex","Embarked","Ticket","PassengerId"], axis = 1, inplace=True)
```

```python
train.head()
```

```python
plt.figure(figsize = (18,6))
sns.heatmap(train)
```

# Spliting Data

```python
y = train["Survived"]
X = train.drop("Survived", axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

# Logistic Regression Algorithm 

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
prediction = logmodel.predict(X_test)
```

```python
print(classification_report(y_test,prediction))
```

```python
print(confusion_matrix(y_test,prediction))
```

# KNN algorithm

```python
# the Best K value
error_rate = []
for i in range(1,40):
    knnmodel = KNeighborsClassifier(n_neighbors=i)
    knnmodel.fit(X_train,y_train)
    knn_prediction = knnmodel.predict(X_test)
    error_rate.append(np.mean(y_test-knn_prediction))

#The Elbow graph
plt.figure(figsize=(15,6))
plt.plot(range(1,40), error_rate, color = "green", marker='o', linestyle='dashed', linewidth=2, markersize=12 , markerfacecolor='red')
plt.title("Error Rate versus K value")
plt.xlabel("K value")
plt.ylabel("Error Rate")  
```

```python
# the K = 3
knnmodel = KNeighborsClassifier(n_neighbors=3)
knnmodel.fit(X_train,y_train)
knn_prediction = knnmodel.predict(X_test)
```

```python
#Confusion Matrix
print(confusion_matrix(y_test,knn_prediction))
```

```python
#Classification Report
print(classification_report(y_test, knn_prediction))
```
